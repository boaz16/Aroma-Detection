```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
import serial
from joblib import dump, load
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv(r'C:\Users\bovas\Desktop\Main Project\Aroma Detection\data.csv')

y = data['tag']
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Load the pre-trained model
model = load_model(r"C:\Users\bovas\Desktop\Main Project\Aroma Detection\ann_model")

# Load the saved scaler from the training phase
scaler = StandardScaler()
scaler_path = (r'C:\Users\bovas\Desktop\Main Project\Aroma Detection\scaler.pkl')
scaler = load(scaler_path)

# Set the serial port 
ser = serial.Serial('COM6', 115200)

while True:
    arduino_data = ser.readline().decode('utf-8').strip().split(',')

    if arduino_data:
        sensor_values = [float(value) for value in arduino_data]
        sensor_df = pd.DataFrame([sensor_values], columns=['voc1', 'no2', 'eth', 'co'])  
        scaled_input = scaler.transform(sensor_df)

        print("\n")
        print("Raw Arduino Data:", arduino_data)
        print("Scaled Input:", scaled_input)

        # Make predictions
        predictions = model.predict(scaled_input)
        print("Raw Predictions:", predictions)

        # Decode the predicted labels
        decoded_labels = label_encoder.inverse_transform(np.argmax(predictions, axis=1))
        print("Predicted Labels:", decoded_labels)
```
